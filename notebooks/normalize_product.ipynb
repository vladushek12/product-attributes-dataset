## Функции для получения метрик схожести.

### Перекрытие терминов.

```python
def preprocess_text(text: str) -> Set[str]:
    """
    Предобработка текста: удаление знаков препинания, приведение к нижнему регистру,
    разбиение на слова и возврат множества уникальных терминов.
    """
    # Удаляем знаки препинания и приводим текст к нижнему регистру
    text = re.sub(r'[^\w\s]', '', text.lower())
    # Разбиваем текст на слова
    words = text.split()
    # Возвращаем множество уникальных слов
    return set(words)

def term_overlap_metric(text1: str, text2: str) -> float:
    """
    Вычисление метрики перекрытия терминов между двумя текстами.
    
    :param text1: Первый текст
    :param text2: Второй текст
    :return: Значение метрики перекрытия терминов (от 0 до 1)
    """
    # Получаем множества уникальных терминов для обоих текстов
    terms1 = preprocess_text(text1)
    terms2 = preprocess_text(text2)
    
    # Находим пересечение множеств (общие термины)
    common_terms = terms1.intersection(terms2)
    
    # Находим объединение множеств (все уникальные термины)
    all_terms = terms1.union(terms2)
    
    # Вычисляем метрику как отношение размера пересечения к размеру объединения
    if len(all_terms) == 0:
        # Если оба текста пустые, возвращаем 0
        return 0.0
    return len(common_terms) / len(all_terms)
```
