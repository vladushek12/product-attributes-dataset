Анализ датасета по категориям

```python
import glob
from pathlib import Path
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
json_filenames = set(filename for filename in glob.glob("cleaned\*.json"))
json_temp_files = set(filename for filename in glob.glob("cleaned\*_temp.json"))

json_filenames -= json_temp_files
```

```python
len(json_temp_files)
```

```python
len(json_filenames)
```

```python
# Читаем файлы, оставляя только нужные колонки
dfs = [pd.read_json(file) for file in json_filenames]

# Объединяем в один DataFrame
df = pd.concat(dfs, ignore_index=True).sort_values(by="count")
```

```python
df
```

```python
df[["count"]].sum()
```

```python
import random

x = 2000
df_dict = df[x:].to_dict()

from collections import Counter

cnt = Counter([df_dict["category_name"][i] for i in random.choices(list(df_dict["category_name"].keys()), k = 100)])
cnt
```

```python
fig = px.bar(df, x="category_name", y="count", title="Количество товаров по категориям")

fig.show()
```

```python
products_dfs = [pd.read_json(file) for file in json_filenames]

products_dfs = pd.concat(products_dfs, ignore_index=True)
```

```python
products_dfs['name'] = products_dfs['products'].apply(lambda x: x['title'])
```

```python
# products_dfs[['name']]

# Найти строки, содержащие "apple" (где угодно в строке)
filtered_df = products_dfs[products_dfs['name'].str.contains(r'Уценка', regex=True)]
filtered_df.shape[0]
```

```python
products_dfs_cleaned = products_dfs[~products_dfs['name'].str.contains(r'Уценка', regex=True)]
```

```python
products_dfs_cleaned.shape[0]
```

```python
products_dfs_cleaned.to_json("data.json", force_ascii=False)
```

```python
import json
import re


for filename in json_filenames:
    name = Path(filename).stem

    with open(filename, encoding='utf-8') as file:
        file_data = json.load(file)

    file_data["products"] = [product for product in file_data["products"] if not re.search(r'Уценка', product['title'])]

    # for product in file_data['products']:
    #     if re.search(r'Уценка', product['title']):
    #         print(product['title'])

    with open(f"cleaned\{name}.json", "w", encoding='utf-8') as file:
        json.dump(file_data, file, ensure_ascii=False)
```
